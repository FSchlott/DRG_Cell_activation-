```python
import matplotlib.pyplot as plt
import numpy as np
import os

from skimage import filters 
from skimage.morphology import disk
from skimage.color import hsv2rgb
from tifffile import imsave
from PIL import Image
```

# Erstellen und Speichern der Iba1 Maske für alle Bilder einer Färbung

```python
def Iba1_mask (img, saveto , filename):
    enhanced = filters.unsharp_mask (img , radius = 20.0 , amount = 2 ) #sharpens the image and reduces background 
    otsu_thr = filters.threshold_otsu (enhanced) #set threshold
    otsu = enhanced > otsu_thr #otsu as array 
    mask = filters.median (otsu , disk (2)) #reduce noise 
    
    im = Image.fromarray (mask)
    im.save (saveto + "/" + filename.replace ("tif" ,"png" ))
    print ("Titel:\t" , saveto + "/" + filename.replace ("tif" ,"png" ))
```

```python
rootdir = r"D:\Immunfluoreszenz\rDRG_CCI_d7\Tifs\1-NF_GFAP_Iba1"  
for dirpath , dirnames , filenames in os.walk (rootdir) : 
    print (dirpath)
    for dirname in dirnames:
        if dirname == "iba1": 
            subdir = os.path.join (dirpath, dirname)
            saveto = subdir.replace ("iba1" , "iba1_pred/masks") 
            print ("\nOrdner\t\t", subdir, "\nSpeicherort\t" , saveto)
            try: #makes directory if it doesnt exist
                os.makedirs (saveto)
            except:
                pass
            for dirpath , dirnames , filenames in os.walk (subdir):
                for filename in filenames:
                    img_f = os.path.join (dirpath, filename)
                    print ("Bild:\t" , img_f)
                    img = plt.imread (img_f) 
                    Iba1_mask (img, saveto , filename) #generates and saves segmentation
            print ("\n")
```
